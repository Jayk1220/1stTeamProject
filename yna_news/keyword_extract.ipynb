```python
import tensorflow as tf
from transformers import BertTokenizer, TFBertModel
from kobert_tokenizer import KoBertTokenizer
# 1. 토크나이저 및 모델 로드
# monologg님의 kobert 모델은 Hugging Face에서 가장 널리 쓰이는 버전입니다.
tokenizer = KoBertTokenizer.from_pretrained('monologg/kobert')
model = TFBertModel.from_pretrained('monologg/kobert', from_pt=True) # PyTorch 가중치를 TF로 변환

# 2. 텍스트 토큰화 예시
text = "오늘 뉴스 데이터 분석은 매우 흥미롭네요."
encoded_input = tokenizer(text, return_tensors='tf', padding='max_length', max_length=64, truncation=True)

# 3. 모델 추론 (벡터 추출)
output = model(encoded_input)
# 'last_hidden_state'는 각 토큰의 의미를 담은 벡터입니다.
# 'pooler_output'은 문장 전체의 의미를 담은 하나의 벡터([CLS] 토큰)입니다.
sentence_vector = output.pooler_output

print(f"문장 벡터 크기: {sentence_vector.shape}") # (1, 768)
```
