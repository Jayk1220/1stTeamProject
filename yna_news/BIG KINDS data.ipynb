```python
merged = (
    big_filter['통합 분류1'].fillna('') + ' ' +
    big_filter['통합 분류2'].fillna('') + ' ' +
    big_filter['통합 분류3'].fillna('')
)

conditions = [
    merged.str.contains('자동차'),
    merged.str.contains('부동산'),
    merged.str.contains('의료')
]

choices = ['자동차', '건설', '헬스케어']

big_filter['산업'] = np.select(conditions, choices, default=np.nan)
```

```python
big_filter['산업'].isna().sum()
```

```python
bigkinds = big_filter.drop(columns=['통합 분류1','통합 분류2','통합 분류3'])
bigkinds
```

```python
bigkinds['일자'] = pd.to_datetime(bigkinds['일자'], format='%Y%m%d')

bigkinds = bigkinds[['일자','제목','특성추출(가중치순 상위 50개)','URL','산업']]
bigkinds
```

```python
# 컬럼 이름 변경

bigkinds = bigkinds.rename(columns={
    '일자': 'date',
    '제목': 'title',
    '특성추출(가중치순 상위 50개)': 'keywords'
})
```
