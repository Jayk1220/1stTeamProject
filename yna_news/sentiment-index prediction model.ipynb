```python
import pandas as pd
```

```python
sent = pd.read_pickle('daily_sentiment.pickle')
index_ = pd.read_csv('data/주식한달데이터.csv')
```

```python
index_['날짜'] = pd.to_datetime(index_['날짜'], format='%Y%m%d')
index_['날짜'] = index_['날짜'].dt.strftime('%Y-%m-%d')
```

```python
index_
```

```python
# 종가
pivot_index_price = index_.pivot(index='날짜', columns='종목명', values='종가(백만원)')
pivot_index_price
```

```python
# 거래량
pivot_index_volumn = index_.pivot(index='날짜', columns='종목명', values='거래량(천주)')
pivot_index_volumn
```

```python
# 1. 아까 만든 pivot_df를 다시 long format으로 (melt 사용)
df_melted = sent.reset_index().melt(id_vars='date', var_name='industry', value_name='sentiment_score')
df_melted
```

```python
# 2. 산업별 기사 비율 계산 (이건 원본 데이터에서 따로 계산해서 합쳐야 함)
# df_melted['article_ratio'] = ... 

# 3. 주가 및 거래량 데이터 Merge
# 'date'와 'industry' 두 가지를 기준으로 병합(Merge)합니다.
final_df = pd.merge(df_melted, index_, on=['date', 'industry'], how='left')

# 4. Target(정답) 만들기: 주가 데이터를 1일, 2일 뒤로 밀어서 생성
final_df['D+1 종가'] = final_df.groupby('industry')['종가(백만원)'].shift(-1)
final_df['D+2 종가'] = final_df.groupby('industry')['종가(백만원)'].shift(-2)
final_df['D+1 거래량'] = final_df.groupby('industry')['종가(백만원)'].shift(-1)
final_df['D+2 거래량'] = final_df.groupby('industry')['종가(백만원)'].shift(-2)
```
